1. Reading Data from file

```python
df= spark.read.csv("WorldCupPlayers.csv", header = True)
df.show()
```

2. Schema of Dataframe

```python
df.printSchema()
```

3. count of the number of Rows and Columns of Dataframe.

```python
column=len(df.columns)
print(column)
row=df.count()
print(row)
```

4. Show the player name and coach name

```python
df.select("PlayerName","CoachName").show()
```

5. Show the records where Match ID must be equal to 1096 and calculate how many records/rows are there in the filtered output.

```python
df.filter(col("MatchID")==1096).show()
```

6. order the data by match id

```python
from pyspark.sql.functions import *
df1=df.orderBy(df.MatchID.asc())
df1.show()
```

7. Show the records of position C and event is G40'.

```python
df3=df.filter((df.Position=="C") & (df.Event=="G40'"))
df3.show()
```

8. Show the unique records with respect to match id.

```python
df6=df.dropDuplicates(["MatchID"])
df6.show()
```

9. Write a sql query to select player name and coach name from dataframe.

```python
df.createOrReplaceTempView("WorldCupPlayers")
spark.sql("SELECT PlayerName,CoachName FROM  WorldCupPlayers").show()
```

10. add a new column year which contains constant value "1930".

```python
df.withColumn("Year",lit("1930")).show()
```

11. Create a new column IsGoalEvent, check if Event column contains 'G'
      1. The new column should contain Y if Event column contains 'G'
      2. The new column should contain N if Event column does not contain 'G'


```python
df7=df.withColumn("IsGoalEvent",when(col("Event").contains("G"),"Y").otherwise("N"))
df7.show()
```

### 2. for mtcars file:

1. Find the top five records of mpg attribute.

```python
mt.select("mpg").show(5)
```

2. Find out the records where the value of mpg is less than 18.

```python
mt.filter(col("mpg") < 18).show()
```

3. Add a new column named as "wtTon" to dataframe with values of weight
column multiplied by 0.45.

```python
mt.withColumn("wtTon",col("wt")*0.45).show()
```

4. Compute the average weight of cars by their cylinders.


```python
mt.groupBy("cyl").agg(avg("wt").alias("avg_weight")).show()
```

5. Write a sql query to find gears where the value of cylinder lies
between 4 and 9.

```python
mt.createOrReplaceTempView("mtg")
spark.sql("SELECT gear FROM  mtg WHERE cyl BETWEEN 4 AND 9").show()
```
